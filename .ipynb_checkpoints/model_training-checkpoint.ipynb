# TypeProfile: Model training notebook

### Dependencies

```python
import pandas as pd
import xgboost as xgb
import seaborn as sns


import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import os

DATA_FILEPATH = os.path.join("Data","DSL-StrongPasswordData.csv")
FINAL_MDOEL_OUT_FILEPATH = os.path.join("Models","ty-v1.joblib")
```

### Loading and Preparing the Data

```python
df = pd.read_csv(DATA_FILEPATH)
df.head()
```

```python
def DataPreprocessing(data):
    data.drop(columns = ["sessionIndex","rep"],inplace = True)
    return data

preprocessed_df = DataPreprocessing(df)
preprocessed_df.head()
```

```python
preprocessed_df.describe()
```

```python
scaler = StandardScaler()
pca = PCA(n_components=3)

X = preprocessed_df.drop(columns = ["subject"]).values
X_scaled = scaler.fit_transform(X)
X_pca = pca.fit_transform(X_scaled)
```

### Visualize the datapoints

```python
pca_df = pd.DataFrame({"pca-1":X_pca[:,0],
                       "pca-2":X_pca[:,1],
                       "pca-3":X_pca[:,2],
                      "subject":preprocessed_df["subject"]})
pca_df.head()
```

```python
sns.set_theme(style="darkgrid")
sns.scatterplot(data = pca_df,
                x = "pca-1",y = "pca-2",legend = False)
plt.title("PCA Plot of Keystroke Features")
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['pca-1'], pca_df['pca-2'], pca_df['pca-3'])

plt.title('3D PCA of Keystroke Features')
plt.show()
```
