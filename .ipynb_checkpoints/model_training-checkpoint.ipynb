# TypeProfile: Model training notebook

### Dependencies

```python
import pandas as pd
import xgboost as xgb
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import os

DATA_FILEPATH = os.path.join("Data","DSL-StrongPasswordData.csv")
FINAL_MDOEL_OUT_FILEPATH = os.path.join("Models","ty-v1.joblib")
```

### Loading and Preparing the Data

```python
df = pd.read_csv(DATA_FILEPATH)
df.head()
```

```python
def DataPreprocessing(data):
    ds = data.drop(columns = ["subject","sessionIndex","rep"])
    return ds

preprocessed_df = DataPreprocessing(df)
preprocessed_df.head()
```

```python
preprocessed_df.describe()
```

```python
scaler = StandardScaler()
pca = PCA(n_components=3)

X = preprocessed_df.values
X_scaled = scaler.fit_transform(X)
X_pca = pca.fit_transform(X_scaled)
```

### Visualize the datapoints

```python
pca_df = pd.DataFrame({"pca-1":X_pca[:,0],
                       "pca-2":X_pca[:,1],
                       "pca-3":X_pca[:,2],
                      "subject":df["subject"]})
pca_df.head()
```

```python
sns.set_theme(style="darkgrid")
sns.scatterplot(data = pca_df,
                x = "pca-1",y = "pca-2",legend = False)
plt.title("PCA Plot of Keystroke Features")
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['pca-1'], pca_df['pca-2'], pca_df['pca-3'])

plt.title('3D PCA of Keystroke Features')
plt.show()
```

### Model Training


```python
knn = KMeans(n_clusters = 3,random_state=4)

clusters = knn.fit_predict(X_scaled)
clusters
```

```python
pca_df["cluster"] = clusters

sns.set_theme(style="darkgrid")
sns.scatterplot(data = pca_df,
                x = "pca-1",y = "pca-2",hue = "cluster",legend = True)
plt.title("PCA Plot of Keystroke Features")
plt.show()
```

```python
preprocessed_df["cluster"] = clusters
preprocessed_df.head()
```

```python
preprocessed_df.groupby("cluster").mean()
```

```python
def ClusterLabel(x):
    if x==0:
        label = "medium"
    elif x==1:
        label = "fast"
    else:
        label = "slow"
    return label
```
